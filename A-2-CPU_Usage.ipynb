```python
import psutil
import time

def monitor_cpu_usage(interval = 1):
    print("Monitoring CPU usage at every second. Press Ctrl+C to stop monitoring")
    try:
        while True:
            cpu_percent = psutil.cpu_percent(interval=interval)
            print(f"CPU Usage : {cpu_percent}%")

            if cpu_percent > 90:
                print(f"Critical Alert, CPU usage above 90% : {cpu_percent}%")
            elif cpu_percent > 85 and cpu_percent < 90:
                print(f"Warning Alert, CPU usage above 85% : {cpu_percent}%")
    except KeyboardInterrupt:
        print("\nMonitoring stopped by user (Keyboard Interrupt).")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")

if __name__ == "__main__":
    monitor_cpu_usage(interval=1)
```
