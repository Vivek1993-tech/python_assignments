```python
#PYTHON CODE TO CHECK PASSWORD STRENGTH
import re
def check_password_strength():
    reg = ['!','@','#','$','%','^','&','*']
    passwd = input("Please enter the password")
    if(len(passwd) < 8):
        print("Password length should be greater than or equal to 8 characters")
    if not any(i.isupper() for i in passwd):
        print("Password must contain atleast one capital letter")
    if not any(i.islower() for i in passwd):
        print("Password must contain atleast one small letter")
    if not any(i.isdigit() for i in passwd):
        print("Password must contain atleast one number digit")
    if not any(i in reg for i in passwd):
        print("Password must contain atleast one symbol like !,@,#,$,%,^,&,*")
    else:
        print("Success!! Your password has been saved")
check_password_strength()
```
